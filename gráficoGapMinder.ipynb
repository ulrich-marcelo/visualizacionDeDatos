```python
# -------------------------------
# 1. Align years (common columns)
# -------------------------------
# Assume format: first column = "country" (or "Country Name"), others are years
inflation_years = set(inflation.columns[1:])
lifeexp_years = set(lifeexp.columns[1:])

common_years = sorted(list(inflation_years & lifeexp_years))

inflation = inflation[["country"] + common_years]
lifeexp = lifeexp[["country"] + common_years]
```

```python
# -------------------------------
# 2. Keep only common countries
# -------------------------------
common_countries = set(inflation["country"]) & set(lifeexp["country"])

inflation = inflation[inflation["country"].isin(common_countries)].reset_index(drop=True)
lifeexp = lifeexp[lifeexp["country"].isin(common_countries)].reset_index(drop=True)
```

```python
# -------------------------------
# 3. Handling NAs
# -------------------------------
# Recommendation: Keep NAs, don't impute them
# Explanation:
# - 0 inflation has real meaning
# - If you impute, you distort trends
# - Best for animation: let NAs mean "country not shown yet"

# -------------------------------
# 4. Build Continents.csv
# -------------------------------
# We can use pycountry_convert or a manual mapping

import pycountry_convert as pc

def get_continent(country_name):
    try:
        # Convert country name to alpha-2
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        # Get continent code
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        # Map to human-readable continent name
        continent_map = {
            "AF": "Africa",
            "AS": "Asia",
            "EU": "Europe",
            "NA": "North America",
            "SA": "South America",
            "OC": "Oceania",
            "AN": "Antarctica"
        }
        return continent_map[continent_code]
    except:
        return "Unknown"

# Build continent dataframe
countries = sorted(list(common_countries))
continent_df = pd.DataFrame({
    "country": countries,
    "continent": [get_continent(c) for c in countries]
})

# Save continents file
continent_df.to_csv("Continents.csv", index=False)

print("Datasets aligned!")
print("Inflation shape:", inflation.shape)
print("LifeExp shape:", lifeexp.shape)
print("Continents file created.")
```

```python
import pandas as pd

# -------------------------------
# Reshape to long format
# -------------------------------
# Inflation
inflation_long = inflation.melt(id_vars="country", var_name="year", value_name="inflation")

# Life expectancy
lifeexp_long = lifeexp.melt(id_vars="country", var_name="year", value_name="lifeExp")

# -------------------------------
# Merge both datasets
# -------------------------------
df = pd.merge(lifeexp_long, inflation_long, on=["country", "year"], how="inner")

# -------------------------------
# Add continents
# -------------------------------
df = pd.merge(df, continent_df, on="country", how="left")

# -------------------------------
# Clean up types
# -------------------------------
df["year"] = df["year"].astype(int)

# Reorder columns
df = df[["country", "continent", "year", "lifeExp", "inflation"]]

# Save final dataframe
df.to_csv("final_dataset.csv", index=False)

print("Final dataset shape:", df.shape)
print(df.head())
```
