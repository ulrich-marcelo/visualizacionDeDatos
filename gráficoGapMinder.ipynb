```python
import pandas as pd

lifeExp = pd.read_csv('data/lifeExp.csv')
inflation = pd.read_csv('data/inflation_annual_percent.csv')
```

```python
#Me quedo desde el 1961 hasta 2024
print(lifeExp.columns)
#
for year in range(1800, 1961):
    lifeExp = lifeExp.drop(columns=str(year))
for year in range(2025, 2101):
    lifeExp = lifeExp.drop(columns=str(year))
print(lifeExp.columns)
```

```python
# transformo los datos de inflacion (hay 12k = 12000, relacionados a procesos hiperinflacionarios)

def parse_value(x):
    if pd.isna(x):
        return float('nan')
    x = str(x).strip()
    if 'k' in x:
        return float(x.replace('k', '')) * 1000
    try:
        return float(x)
    except ValueError:
        return float('nan')

# aplico solo a años
year_columns = [col for col in inflation.columns if col != 'country']
for col in year_columns:
    inflation[col] = inflation[col].apply(parse_value)
```

```python
# 1. Mantengo los paises comunes, por las dudas

common_countries = set(inflation["country"]) & set(lifeExp["country"])

inflation = inflation[inflation["country"].isin(common_countries)].reset_index(drop=True)
lifeExp = lifeExp[lifeExp["country"].isin(common_countries)].reset_index(drop=True)
```

```python
continent_df = pd.read_csv("data/continents.csv")
```

```python
# Armo el dataframe grande.

# Inflation
inflation_long = inflation.melt(id_vars="country", var_name="year", value_name="inflation")

# Life expectancy
lifeexp_long = lifeExp.melt(id_vars="country", var_name="year", value_name="lifeExp")


# Merge both datasets
df = pd.merge(lifeexp_long, inflation_long, on=["country", "year"], how="inner")


# Add continents
df = pd.merge(df, continent_df, on="country", how="left")


# Clean up types
df["year"] = df["year"].astype(int)

# Reorder columns
df = df[["country", "continent", "year", "lifeExp", "inflation"]]

# Save final dataframe
df.to_csv("data/inflationVsLifeExp.csv", index=False)

print("Final dataset shape:", df.shape)
print(df.head())
```
